```python
import numpy as np
import pandas as pd
```

```python
salaries = pd.read_csv('Salaries.csv')
```

```python
salaries.head()
```

```python
salaries.info()
```

```python
salaries.isnull().sum()
```

```python
#deleting the Notes column because it has no observations

salaries.drop('Notes', inplace=True, axis=1)
```

```python
#converting numerical columns to float datatype rather than object datatype

salaries['BasePay'] = pd.to_numeric(salaries['BasePay'],errors = 'coerce')
salaries['OvertimePay'] = pd.to_numeric(salaries['OvertimePay'],errors = 'coerce')
salaries['OtherPay'] = pd.to_numeric(salaries['OtherPay'],errors = 'coerce')
salaries['Benefits'] = pd.to_numeric(salaries['Benefits'],errors = 'coerce')
```

```python
#check datatype after conversion

salaries.info() 
```

```python
salaries.tail()
```

```python
#converting the 'not provided' values to null values

salaries.replace('Not provided', np.nan, inplace=True)
```

```python
#checking for unlogical values like negative total pay

salaries.sort_values(by = 'TotalPay',ascending = True)
```

```python
salaries.drop([148653], inplace=True, axis=0)
```

```python
salaries.tail()
```

**What is the job title of GARY JIMENEZ ?**

```python
salaries[salaries['EmployeeName'] == 'GARY JIMENEZ']['JobTitle']
```

**How much does GARY JIMENEZ make (including benefits) ?**

```python
salaries[salaries['EmployeeName'] == 'GARY JIMENEZ']['TotalPayBenefits']
```

**What is the name of highest paid person (including benefits)?**

```python
salaries[salaries['TotalPayBenefits'] == salaries['TotalPayBenefits'].max()]['EmployeeName']
```
